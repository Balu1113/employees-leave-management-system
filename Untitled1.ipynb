```python
sentences = ["Apply a casual leave for Balaji from 2025-10-04 to 2025-10-05",
    "Please process casual leave for Abhinesh between 2025-10-04 and 2025-10-05",
    "Submit a casual leave request for Sumanth from Oct 4th to Oct 5th, 2025",
    "Revanth needs casual leave starting 2025-10-04 till 2025-10-05",
    "Log casual leave for Deepak from 4th October to 5th October 2025",
    "Kindly mark Surya on casual leave from 2025-10-04 to 2025-10-05",
    "Dheeraj wants a casual leave from 2025-10-04 until 2025-10-05",
    "Ravi will be on casual leave from 4/10/2025 to 5/10/2025",
    "Sita applied for casual leave between 2025-10-04 and 2025-10-05",
    "John has requested casual leave from 2025-10-04 to 2025-10-05",

    # Sick Leave
    "Apply sick leave for Balaji from 2025-10-04 to 2025-10-05",
    "Please apply a sick leave for Abhinesh between 4th and 5th October 2025",
    "Sumanth is taking sick leave from 2025-10-04 till 2025-10-05",
    "Revanth needs to apply sick leave from Oct 4â€“5, 2025",
    "Log a sick leave for Deepak from 2025-10-04 to 2025-10-05",
    "Surya requested sick leave for 4th and 5th October 2025",
    "Dheeraj has applied for sick leave from 2025-10-04 till 2025-10-05",
    "Ravi should be marked on sick leave between 4â€“5 Oct 2025",
    "Please mark Sita on sick leave from 2025-10-04 to 2025-10-05",
    "Johnâ€™s sick leave runs from October 4th to 5th, 2025",

    # Annual Leave
    "Apply annual leave for Balaji from 2025-10-04 to 2025-10-05",
    "Please apply annual leave for Abhinesh between 4/10/2025 and 5/10/2025",
    "Submit an annual leave request for Sumanth from 2025-10-04 till 2025-10-05",
    "Revanth is on annual leave from 2025-10-04 to 2025-10-05",
    "Deepak needs annual leave for 4th and 5th October 2025",
    "Surya will take annual leave from Oct 4th to Oct 5th, 2025",
    "Apply annual leave on behalf of Dheeraj from 2025-10-04 to 2025-10-05",
    "Mark Ravi on annual leave between 2025-10-04 and 2025-10-05",
    "Sita has planned annual leave from 4â€“5 October 2025",
    "Johnâ€™s annual leave covers 2025-10-04 to 2025-10-05",

    # Earned Leave
    "Apply earned leave for Balaji from 2025-10-04 to 2025-10-05",
    "Abhinesh needs earned leave from Oct 4th to Oct 5th, 2025",
    "Please mark Sumanth on earned leave between 2025-10-04 and 2025-10-05",
    "Submit earned leave for Revanth from 2025-10-04 to 2025-10-05",
    "Log Deepakâ€™s earned leave from 4/10/2025 to 5/10/2025",
    "Surya applied for earned leave from October 4â€“5, 2025",
    "Apply earned leave for Dheeraj from 2025-10-04 till 2025-10-05",
    "Ravi will take earned leave from 4th October to 5th October 2025",
    "Sita needs earned leave from 2025-10-04 to 2025-10-05",
    "John requested earned leave between Oct 4th and Oct 5th, 2025",

    # Marriage & Vacation Leave
    "Apply marriage leave for Balaji from 2025-10-04 to 2025-10-05",
    "Please process marriage leave for Abhinesh between 4th and 5th October 2025",
    "Mark Sumanth on marriage leave from 2025-10-04 till 2025-10-05",
    "Revanth applied for marriage leave from Oct 4â€“5, 2025",
    "Deepak needs marriage leave from 2025-10-04 to 2025-10-05",
    "Surya is on marriage leave from 2025-10-04 to 2025-10-05",
    "Apply vacation leave for Dheeraj from 2025-10-04 till 2025-10-05",
    "Please mark Ravi on vacation leave from 2025-10-04 to 2025-10-05",
    "Sita requested vacation leave for 4thâ€“5th October 2025",
    "John has vacation leave from 2025-10-04 to 2025-10-05"]

# Define employee names and leave types
employee_names = ["Balaji", "Abhinesh", "Sumanth", "Revanth", "Deepak", "Surya", "Dheeraj", "Ravi", "Sita", "John"]
leave_types = ["casual leave", "sick leave", "annual leave", "earned leave", "marriage leave", "vacation leave"]

# Extract employee names and leave types from sentences
extracted_employee_names = []
extracted_leave_types = []
for sentence in sentences:
    found_name = False
    for name in employee_names:
        if name in sentence:
            extracted_employee_names.append(name)
            found_name = True
            break
    if not found_name:
        extracted_employee_names.append("Unknown") # Handle cases where name is not found

    found_leave = False
    for leave_type in leave_types:
        if leave_type in sentence:
            extracted_leave_types.append(leave_type)
            found_leave = True
            break
    if not found_leave:
        extracted_leave_types.append("Unknown") # Handle cases where leave type is not found


from_dates = ["2025-10-04"] * len(sentences)
to_dates = ["2025-10-05"] * len(sentences)
```

```python
tokenizer = Tokenizer(lower=True, oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
X = tokenizer.texts_to_sequences(sentences)
max_len = max(len(seq) for seq in X)
X = pad_sequences(X, padding='post', maxlen=max_len)
vocab_size = len(tokenizer.word_index) + 1

emp_encoder = LabelEncoder()
leave_encoder = LabelEncoder()
from_encoder = LabelEncoder()
to_encoder = LabelEncoder()

# Fit and transform on the extracted lists
y_emp = emp_encoder.fit_transform(extracted_employee_names)
y_leave = leave_encoder.fit_transform(extracted_leave_types)
y_from = from_encoder.fit_transform(from_dates)
y_to = to_encoder.fit_transform(to_dates)


# One-hot encode for classification for employee and leave type
y_emp = to_categorical(y_emp)
y_leave = to_categorical(y_leave)
# Keep dates as label encoded for regression
# y_from = to_categorical(y_from)
# y_to = to_categorical(y_to)
```

```python
inp = Input(shape=(max_len,))
x = Embedding(vocab_size, 64)(inp)
x = Conv1D(128, 3, activation='relu')(x)
x = GlobalMaxPooling1D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)

# Separate outputs
emp_out = Dense(y_emp.shape[1], activation='softmax', name='emp_out')(x)
leave_out = Dense(y_leave.shape[1], activation='softmax', name='leave_out')(x)
# Use linear activation for date outputs and a single output neuron
from_out = Dense(1, activation='linear', name='from_out')(x)
to_out = Dense(1, activation='linear', name='to_out')(x)

model = Model(inputs=inp, outputs=[emp_out, leave_out, from_out, to_out])

model.compile(
    optimizer='adam',
    loss={'emp_out': 'categorical_crossentropy', 'leave_out': 'categorical_crossentropy', 'from_out': 'mean_squared_error', 'to_out': 'mean_squared_error'},
    metrics={'emp_out': 'accuracy', 'leave_out': 'accuracy', 'from_out': 'mae', 'to_out': 'mae'},
    jit_compile=True, # Enable JIT compilation for potential performance improvement
    run_eagerly=False # Disable eager execution for potential performance improvement
)

model.summary()
```

```python
history = model.fit(
    X,
    [y_emp, y_leave, y_from, y_to],
    epochs=200,
    batch_size=2,
    verbose=0
)

print("\n Model trained successfully!")
```

```python
def predict_leave(sentence):
    seq = tokenizer.texts_to_sequences([sentence])
    seq = pad_sequences(seq, maxlen=max_len, padding='post')

    pred_emp, pred_leave, pred_from, pred_to = model.predict(seq, verbose=0)

    emp_pred = emp_encoder.inverse_transform([np.argmax(pred_emp)])[0]
    leave_pred = leave_encoder.inverse_transform([np.argmax(pred_leave)])[0]
    from_pred = from_encoder.inverse_transform([np.argmax(pred_from)])[0]
    to_pred = to_encoder.inverse_transform([np.argmax(pred_to)])[0]

    return {
        "employee_name": emp_pred,
        "leave_type": leave_pred,
        "from_date": from_pred,
        "to_date": to_pred
    }
```

```python
test_sentence = "Revanth needs to apply sick leave from Oct 7â€“8, 2025"
result = predict_leave(test_sentence)

print("\nğŸ”¹ Prediction:")
print(f"Employee Name : {result['employee_name']}")
print(f"Leave Type    : {result['leave_type']}")
print(f"From Date     : {result['from_date']}")
print(f"To Date       : {result['to_date']}")
```
