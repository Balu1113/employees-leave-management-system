```python
sentences = [
    # --- CASUAL LEAVE ---
    "Please apply a casual leave for Balaji",
    "Balaji needs casual leave tomorrow",
    "Kindly submit casual leave for Balaji",
    "Mark Balaji as on casual leave today",
    "Apply one day of casual leave for Balaji",

    "Request casual leave for Abhinesh",
    "Apply a casual leave on behalf of Abhinesh",
    "Abhinesh wants casual leave for tomorrow",
    "Please log a casual leave for Abhinesh",
    "Mark Abhinesh on casual leave",

    "Apply casual leave for Sumanth",
    "Sumanth applied for casual leave today",
    "Please record a casual leave for Sumanth",
    "Submit a casual leave request for Sumanth",
    "Log a casual leave for Sumanth",

    "Apply casual leave for Revanth",
    "Revanth took a casual leave today",
    "Please put in a casual leave for Revanth",
    "Revanth wants one day casual leave",
    "Submit casual leave request for Revanth",

    "Deepak applied for casual leave",
    "Please process casual leave for Deepak",
    "Log a casual leave entry for Deepak",
    "Submit a casual leave for Deepak",
    "Mark Deepak as on casual leave",

    "Apply casual leave for Surya",
    "Surya needs casual leave tomorrow",
    "Request casual leave for Surya",
    "Please submit casual leave for Surya",
    "Log casual leave for Surya",

    "Apply casual leave for Dheeraj",
    "Kindly record casual leave for Dheeraj",
    "Dheeraj wants a casual leave for today",
    "Please apply one day casual leave for Dheeraj",
    "Submit casual leave for Dheeraj",

    "Ravi applied for casual leave",
    "Please process casual leave for Ravi",
    "Ravi is taking casual leave today",
    "Apply one day casual leave for Ravi",
    "Submit casual leave for Ravi",

    "Apply casual leave for Sita",
    "Sita requested casual leave today",
    "Please log casual leave for Sita",
    "Apply one day casual leave for Sita",
    "Mark Sita as on casual leave",

    "Apply casual leave for John",
    "Please submit casual leave for John",
    "John took casual leave today",
    "Record casual leave for John",
    "Request casual leave entry for John",

    "Apply casual leave for Mark",
    "Please record casual leave for Mark",
    "Log casual leave for Mark",
    "Submit casual leave for Mark",
    "Mark applied for casual leave",

    # --- SICK LEAVE ---
    "Apply sick leave for Balaji",
    "Please mark Balaji as on sick leave",
    "Submit sick leave for Balaji",
    "Balaji requested sick leave today",
    "Kindly record sick leave for Balaji",

    "Apply sick leave for Abhinesh",
    "Mark Abhinesh as on sick leave",
    "Please process sick leave for Abhinesh",
    "Submit sick leave request for Abhinesh",
    "Record sick leave entry for Abhinesh",

    "Apply sick leave for Sumanth",
    "Sumanth is on sick leave today",
    "Submit sick leave request for Sumanth",
    "Please mark Sumanth on sick leave",
    "Record sick leave for Sumanth",

    "Apply sick leave for Revanth",
    "Revanth needs sick leave tomorrow",
    "Mark Revanth as on sick leave today",
    "Please record sick leave for Revanth",
    "Submit sick leave request for Revanth",

    "Apply sick leave for Deepak",
    "Please mark Deepak as on sick leave",
    "Deepak is taking sick leave today",
    "Submit sick leave for Deepak",
    "Kindly approve sick leave for Deepak",

    "Apply sick leave for Surya",
    "Mark Surya as on sick leave",
    "Please record sick leave for Surya",
    "Submit sick leave for Surya",
    "Surya took sick leave today",

    "Apply sick leave for Dheeraj",
    "Please record sick leave for Dheeraj",
    "Mark Dheeraj on sick leave",
    "Dheeraj applied for sick leave",
    "Submit sick leave request for Dheeraj",

    "Apply sick leave for Ravi",
    "Please record sick leave for Ravi",
    "Ravi applied for sick leave today",
    "Mark Ravi as on sick leave",
    "Request sick leave entry for Ravi",

    "Apply sick leave for Sita",
    "Please mark Sita on sick leave",
    "Sita applied for sick leave",
    "Record sick leave entry for Sita",
    "Submit sick leave request for Sita",

    "Apply sick leave for John",
    "Please log sick leave for John",
    "John applied for sick leave today",
    "Mark John as on sick leave",
    "Submit sick leave request for John",

    "Apply sick leave for Mark",
    "Please log sick leave for Mark",
    "Mark applied for sick leave",
    "Submit sick leave request for Mark",
    "Record sick leave entry for Mark",

    # --- VACATION LEAVE ---
    "Apply vacation leave for Balaji",
    "Balaji is on vacation leave next week",
    "Please process vacation leave for Balaji",
    "Submit vacation leave for Balaji",
    "Log vacation leave entry for Balaji",

    "Apply vacation leave for Abhinesh",
    "Please mark Abhinesh on vacation leave",
    "Abhinesh applied for vacation leave",
    "Submit vacation leave request for Abhinesh",
    "Log vacation leave for Abhinesh",

    "Apply vacation leave for Sumanth",
    "Sumanth is taking vacation leave next week",
    "Please log vacation leave for Sumanth",
    "Submit vacation leave for Sumanth",
    "Record vacation leave entry for Sumanth",

    "Apply vacation leave for Revanth",
    "Please process vacation leave for Revanth",
    "Revanth is taking vacation leave tomorrow",
    "Submit vacation leave for Revanth",
    "Log vacation leave for Revanth",

    "Apply vacation leave for Deepak",
    "Please record vacation leave for Deepak",
    "Deepak applied for vacation leave today",
    "Submit vacation leave request for Deepak",
    "Mark Deepak as on vacation leave",

    "Apply vacation leave for Surya",
    "Surya requested vacation leave",
    "Submit vacation leave request for Surya",
    "Please mark Surya as on vacation leave",
    "Record vacation leave for Surya",

    "Apply vacation leave for Dheeraj",
    "Please mark Dheeraj as on vacation leave",
    "Submit vacation leave request for Dheeraj",
    "Log vacation leave for Dheeraj",
    "Dheeraj applied for vacation leave today",

    "Apply vacation leave for Ravi",
    "Please record vacation leave for Ravi",
    "Ravi applied for vacation leave",
    "Mark Ravi on vacation leave today",
    "Submit vacation leave for Ravi",

    "Apply vacation leave for Sita",
    "Sita is taking vacation leave",
    "Please log vacation leave for Sita",
    "Submit vacation leave for Sita",
    "Record vacation leave for Sita",

    "Apply vacation leave for John",
    "John needs vacation leave next week",
    "Submit vacation leave request for John",
    "Please mark John on vacation leave",
    "Record vacation leave entry for John",

    "Apply vacation leave for Mark",
    "Mark applied for vacation leave",
    "Please record vacation leave for Mark",
    "Submit vacation leave request for Mark",
    "Log vacation leave entry for Mark",

    # --- MARRIAGE LEAVE ---
    "Apply marriage leave for Balaji",
    "Please mark Balaji on marriage leave",
    "Submit marriage leave for Balaji",
    "Balaji applied for marriage leave",
    "Log marriage leave for Balaji",

    "Apply marriage leave for Abhinesh",
    "Please record marriage leave for Abhinesh",
    "Abhinesh needs marriage leave next week",
    "Submit marriage leave request for Abhinesh",
    "Mark Abhinesh as on marriage leave",

    "Apply marriage leave for Sumanth",
    "Sumanth requested marriage leave",
    "Please log marriage leave for Sumanth",
    "Submit marriage leave request for Sumanth",
    "Record marriage leave entry for Sumanth",

    "Apply marriage leave for Revanth",
    "Please mark Revanth on marriage leave",
    "Revanth applied for marriage leave",
    "Submit marriage leave for Revanth",
    "Log marriage leave entry for Revanth",

    "Apply marriage leave for Deepak",
    "Please log marriage leave for Deepak",
    "Submit marriage leave request for Deepak",
    "Mark Deepak as on marriage leave",
    "Record marriage leave for Deepak",

    "Apply marriage leave for Surya",
    "Surya applied for marriage leave",
    "Please mark Surya on marriage leave",
    "Submit marriage leave for Surya",
    "Record marriage leave for Surya",

    "Apply marriage leave for Dheeraj",
    "Please mark Dheeraj on marriage leave",
    "Submit marriage leave for Dheeraj",
    "Log marriage leave for Dheeraj",
    "Record marriage leave entry for Dheeraj",

    "Apply marriage leave for Ravi",
    "Please mark Ravi on marriage leave",
    "Submit marriage leave for Ravi",
    "Ravi applied for marriage leave",
    "Record marriage leave for Ravi",

    "Apply marriage leave for Sita",
    "Please log marriage leave for Sita",
    "Sita applied for marriage leave",
    "Submit marriage leave for Sita",
    "Record marriage leave entry for Sita",

    "Apply marriage leave for John",
    "John requested marriage leave",
    "Please log marriage leave for John",
    "Submit marriage leave request for John",
    "Mark John on marriage leave",

    "Apply marriage leave for Mark",
    "Please mark Mark on marriage leave",
    "Mark applied for marriage leave",
    "Submit marriage leave for Mark",
    "Record marriage leave entry for Mark",

    # --- ANNUAL LEAVE ---
    "Apply annual leave for Balaji",
    "Balaji applied for annual leave",
    "Please process annual leave for Balaji",
    "Submit annual leave request for Balaji",
    "Record annual leave for Balaji",

    "Apply annual leave for Abhinesh",
    "Please log annual leave for Abhinesh",
    "Submit annual leave for Abhinesh",
    "Abhinesh applied for annual leave",
    "Mark Abhinesh as on annual leave",

    "Apply annual leave for Sumanth",
    "Please record annual leave for Sumanth",
    "Sumanth applied for annual leave",
    "Submit annual leave request for Sumanth",
    "Mark Sumanth as on annual leave",

    "Apply annual leave for Revanth",
    "Revanth needs annual leave next week",
    "Please mark Revanth on annual leave",
    "Submit annual leave for Revanth",
    "Log annual leave for Revanth",

    "Apply annual leave for Deepak",
    "Deepak applied for annual leave",
    "Please record annual leave for Deepak",
    "Submit annual leave request for Deepak",
    "Mark Deepak on annual leave",

    "Apply annual leave for Surya",
    "Please log annual leave for Surya",
    "Surya requested annual leave",
    "Submit annual leave for Surya",
    "Record annual leave entry for Surya",

    "Apply annual leave for Dheeraj",
    "Please log annual leave for Dheeraj",
    "Submit annual leave for Dheeraj",
    "Mark Dheeraj on annual leave",
    "Record annual leave entry for Dheeraj",

    "Apply annual leave for Ravi",
    "Ravi applied for annual leave",
    "Please record annual leave for Ravi",
    "Submit annual leave for Ravi",
    "Mark Ravi on annual leave",

    "Apply annual leave for Sita",
    "Please log annual leave for Sita",
    "Sita requested annual leave",
    "Submit annual leave request for Sita",
    "Mark Sita on annual leave",

    "Apply annual leave for John",
    "John applied for annual leave",
    "Please record annual leave for John",
    "Submit annual leave for John",
    "Mark John on annual leave",

    "Apply annual leave for Mark",
    "Please record annual leave for Mark",
    "Mark applied for annual leave",
    "Submit annual leave for Mark",
    "Log annual leave entry for Mark",

    # --- EARNED LEAVE ---
    "Apply earned leave for Balaji",
    "Please log earned leave for Balaji",
    "Balaji requested earned leave",
    "Submit earned leave request for Balaji",
    "Mark Balaji on earned leave",

    "Apply earned leave for Abhinesh",
    "Please record earned leave for Abhinesh",
    "Abhinesh applied for earned leave",
    "Submit earned leave request for Abhinesh",
    "Mark Abhinesh as on earned leave",

    "Apply earned leave for Sumanth",
    "Please log earned leave for Sumanth",
    "Sumanth applied for earned leave",
    "Submit earned leave request for Sumanth",
    "Mark Sumanth as on earned leave",

    "Apply earned leave for Revanth",
    "Revanth requested earned leave",
    "Please log earned leave for Revanth",
    "Submit earned leave for Revanth",
    "Mark Revanth as on earned leave",

    "Apply earned leave for Deepak",
    "Deepak applied for earned leave",
    "Please record earned leave for Deepak",
    "Submit earned leave request for Deepak",
    "Mark Deepak as on earned leave",

    "Apply earned leave for Surya",
    "Please record earned leave for Surya",
    "Submit earned leave for Surya",
    "Surya applied for earned leave",
    "Mark Surya on earned leave",

    "Apply earned leave for Dheeraj",
    "Please record earned leave for Dheeraj",
    "Submit earned leave for Dheeraj",
    "Dheeraj applied for earned leave",
    "Mark Dheeraj on earned leave",

    "Apply earned leave for Ravi",
    "Please log earned leave for Ravi",
    "Submit earned leave for Ravi",
    "Ravi applied for earned leave",
    "Mark Ravi on earned leave",

    "Apply earned leave for Sita",
    "Please log earned leave for Sita",
    "Submit earned leave for Sita",
    "Sita requested earned leave",
    "Mark Sita on earned leave",

    "Apply earned leave for John",
    "Please record earned leave for John",
    "Submit earned leave for John",
    "John applied for earned leave",
    "Mark John on earned leave",

    "Apply earned leave for Mark",
    "Please log earned leave for Mark",
    "Submit earned leave for Mark",
    "Mark applied for earned leave",
    "Record earned leave entry for Mark",
]
# List of known employees and leave types
employee_names = ["Balaji","Abhinesh","Sumanth","Revanth","Deepak","Surya","Dheeraj","Ravi","Sita","John"]
leave_types = ["casual","sick","vacation","marriage","annual","earned"]

# Add "unknown" to handle unseen inputs
if "unknown" not in employee_names:
    employee_names.append("unknown")
if "unknown" not in leave_types:
    leave_types.append("unknown")
```

```python
tokenizer = Tokenizer(lower=True, oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)

X = tokenizer.texts_to_sequences(sentences)
max_len = max(len(x) for x in X)
X = pad_sequences(X, padding="post", maxlen=max_len)
vocab_size = len(tokenizer.word_index) + 1
```

```python
leave_encoder = LabelEncoder()
emp_encoder = LabelEncoder()

leave_encoder.fit(leave_types)
emp_encoder.fit(employee_names)

# Generate labels from text
y_leave_labels = []
y_emp_labels = []

for s in sentences:
    leave_label = "unknown"
    emp_label = "unknown"
    for lt in leave_types:
        if lt in s.lower():
            leave_label = lt
            break
    for emp in employee_names:
        if emp.lower() in s.lower():
            emp_label = emp
            break
    y_leave_labels.append(leave_label)
    y_emp_labels.append(emp_label)

y_leave = leave_encoder.transform(y_leave_labels)
y_emp = emp_encoder.transform(y_emp_labels)

y_leave_one_hot = to_categorical(y_leave, num_classes=len(leave_encoder.classes_))
y_emp_one_hot = to_categorical(y_emp, num_classes=len(emp_encoder.classes_))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train_leave, y_val_leave, y_train_emp, y_val_emp = train_test_split(
    X, y_leave_one_hot, y_emp_one_hot, test_size=0.25, random_state=42
)
```

```python
input_layer = Input(shape=(max_len,))
x = Embedding(input_dim=vocab_size, output_dim=128)(input_layer)
x = Bidirectional(LSTM(128, return_sequences=True))(x)
x = GlobalMaxPooling1D()(x)
x = Dense(128, activation="relu")(x)
x = Dropout(0.4)(x)
x = Dense(64, activation="relu")(x)
leave_output = Dense(len(leave_encoder.classes_), activation="softmax", name="leave_output")(x)
emp_output = Dense(len(emp_encoder.classes_), activation="softmax", name="emp_output")(x)

model = Model(inputs=input_layer, outputs=[leave_output, emp_output])
model.compile(
    optimizer="adam",
    loss={
        "leave_output": "categorical_crossentropy",
        "emp_output": "categorical_crossentropy"
    },
    metrics={
        "leave_output": "accuracy",
        "emp_output": "accuracy"
    }
)
```

```python
history = model.fit(
    X_train,
    [y_train_leave, y_train_emp],
    epochs=175,         # You can increase to 100–150 if dataset is large
    batch_size=4,
    validation_data=(X_val, [y_val_leave, y_val_emp]),
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

def plot_accuracy(history):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Accuracy
    axes[0].plot(history.history['leave_output_accuracy'], label='Leave Train Acc')
    if 'val_leave_output_accuracy' in history.history:
        axes[0].plot(history.history['val_leave_output_accuracy'], label='Leave Val Acc')
    axes[0].plot(history.history['emp_output_accuracy'], label='Emp Train Acc')
    if 'val_emp_output_accuracy' in history.history:
        axes[0].plot(history.history['val_emp_output_accuracy'], label='Emp Val Acc')
    axes[0].set_title('Training & Validation Accuracy')
    axes[0].set_xlabel('Epochs')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()

    # Loss
    axes[1].plot(history.history['loss'], label='Training Loss')
    if 'val_loss' in history.history:
        axes[1].plot(history.history['val_loss'], label='Validation Loss')
    axes[1].set_title('Training & Validation Loss')
    axes[1].set_xlabel('Epochs')
    axes[1].set_ylabel('Loss')
    axes[1].legend()

    plt.tight_layout()
    plt.show()

plot_accuracy(history)
```

```python
model.save("leave_classifier.h5")
print("✅ Model saved successfully as 'leave_classifier.h5'")
eval_results = model.evaluate(X_val, [y_val_leave, y_val_emp], verbose=1)
print(f"\nValidation Loss: {eval_results[0]:.4f}")
print(f"Leave Accuracy: {eval_results[3]:.4f}")
print(f"Employee Accuracy: {eval_results[4]:.4f}")
```

```python
def predict_entities(sentence):
    """
    Predicts the leave type and employee name from a given sentence.

    Args:
        sentence (str): The input sentence.

    Returns:
        dict: A dictionary containing the predicted leave type and employee name.
    """
    seq = tokenizer.texts_to_sequences([sentence])
    seq = pad_sequences(seq, maxlen=max_len, padding="post")

    pred_leave, pred_emp = model.predict(seq)

    leave_pred = leave_encoder.inverse_transform([np.argmax(pred_leave)])[0]
    emp_pred = emp_encoder.inverse_transform([np.argmax(pred_emp)])[0]

    return {"leave_type": leave_pred, "employee_name": emp_pred}
```
